```python
import pandas as pd
```

```python
house_rep_df = pd.read_csv('../data/house_rep_data_to_michigan.csv')
house_rep_df_2 = pd.read_csv('../data/house_rep_data_from_michigan_to_texas.csv')
house_rep_df_3 = pd.read_csv('../data/house_rep_data_from_texas.csv')
```

```python
house_rep_df = pd.concat([house_rep_df,house_rep_df_2,house_rep_df_3]).drop(columns = 'Unnamed: 0')
house_rep_df
```

```python
house_rep_df['District'] = house_rep_df['District'].astype(str)
```

```python
house_rep_df.dtypes
```

```python
house_rep_df = house_rep_df.query("District != '2020'")
```

```python
house_rep_df.to_csv('../data/house_rep_data_final.csv',index = False)
```
